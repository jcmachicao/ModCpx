```
import numpy as np
import pandas as pd
from random import seed
from random import random
import matplotlib.pyplot as plt
```

```
opciones = pd.read_excel('drive/My Drive/2020 data/opciones.xlsx')
```

```
#@title Definición de las probabilidades de transición para 4 estados
P = np.array([
              [0.20, 0.60, 0.05, 0.15],
              [0.70, 0.05, 0.15, 0.10],
              [0.15, 0.00, 0.80, 0.05],
              [0.35, 0.15, 0.15, 0.35]
              ])
```

```
#@title Definición de estado inicial
state = np.array([[0.16, 0.15, 0.13, 0.13]])
stateHist = state
dfStateHist = pd.DataFrame(state)
distr_hist = [[0.16, 0.15, 0.13, 0.13]]
```

```
#@title Generacaión de iteraciones
N_ITER = 10 # Número de iteraciones
for x in range(N_ITER):
  state = np.dot(state, P)
  #print(state)
  stateHist = np.append(stateHist, state, axis=0)
```

```
#@title Gráfica de iteraciones
dfDistrHist = pd.DataFrame(stateHist)
plt.figure(facecolor='white', figsize=(10,8))
plt.plot(dfDistrHist[0], lw=5, alpha=0.8, label='A')
plt.plot(dfDistrHist[1], lw=5, alpha=0.8, label='B')
plt.plot(dfDistrHist[2], lw=5, alpha=0.5, label='C')
plt.plot(dfDistrHist[3], lw=5, alpha=0.5, label='D')


plt.grid(True)
plt.legend()
plt.show()
```

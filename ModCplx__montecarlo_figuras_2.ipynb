```python
def genera_pi():
  matriz = np.random.randint(2, size=300).reshape(15,20)
  xt_mat1 = matriz[2:12, 2:12]
  xt_mat2 = matriz[3:8, 13:18]
  s1 = (xt_mat1[0,5:7].sum() + xt_mat1[1,4:8].sum() + \
       xt_mat1[2:4,3:9].sum() + xt_mat1[4:6,:].sum()) * 2
  s2 = xt_mat2.sum()
  return s1/s2
```

```python
pi_rand = []
for k in range(500):
  g = genera_pi()
  pi_rand.append(g)
  if k < 10:
    print(g)
pi_np = np.array(pi_rand)
print('Promedio: ', pi_np.mean())
```
